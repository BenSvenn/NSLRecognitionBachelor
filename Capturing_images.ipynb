```python
import cv2
import os

# Define the optimal image resolution
optimal_resolution = (640, 480)

# Initialize the webcam
cam = cv2.VideoCapture(0)
cam.set(cv2.CAP_PROP_FRAME_WIDTH, optimal_resolution[0])
cam.set(cv2.CAP_PROP_FRAME_HEIGHT, optimal_resolution[1])

save_dir = './raadata_BS_H_N'

# Define the initial label for the current images
current_label = 'A'

# Create the directory for the initial label if it doesn't exist
label_dir = os.path.join(save_dir, current_label)
if not os.path.exists(label_dir):
    os.makedirs(label_dir)

# Define the region of interest (ROI)
x, y, w, h = 200, 200, 256, 256

# Loop to take pictures
while True:
    # Capture the current frame
    ret, frame = cam.read()
    
    # Check if the frame was read successfully
    if not ret:
        continue

    # Crop the frame to the ROI
    roi = frame[y:y+h, x:x+w]

    # Show the current frame
    cv2.imshow('Webcam', roi)

    # Wait for the user to press a key
    key = cv2.waitKey(1)

    # Check if the user pressed the 's' key
    if key == ord('s'):
        # Generate a file name for the current image
        file_name = current_label + '_' + str(len(os.listdir(label_dir))) + '_U.jpg'
        file_path = os.path.join(label_dir, file_name)

        # Save the current frame to the file
        cv2.imwrite(file_path, roi)

        # Print a message to indicate that the image was saved
        print(f'{file_name} saved.')
    elif key == ord('n'):
        # Increment the current label to the next letter of the alphabet
        current_label = chr(ord(current_label) + 1)

        # Create the directory for the new label if it doesn't exist
        label_dir = os.path.join(save_dir, current_label)
        if not os.path.exists(label_dir):
            os.makedirs(label_dir)

        # Print a message to indicate the new label
        print(f'Label changed to {current_label}.')
    elif key == ord('q'):
        # Release the webcam and close the window
        cam.release()
        cv2.destroyAllWindows()
        break
```
